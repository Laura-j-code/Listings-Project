# EXPLORATORY DATA ANALYSIS
    Goal: summarizing the maine characteristicts of the dataset so as to gain further insights intothe data and formulate hypothese for further analysis
    Activity: Visualization ie. calculating summary statistics, detecting outliers and identifying patterns and correlations

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
sns.countplot(x='room_type', data=df)
plt.title('Distribution of Room Types')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.show()

# tThus more hosts are offering the private rooms
```

```python
sns.histplot(df['price'], bins=30, kde=True)
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Number of Listings')
plt.xlim(0, 500)  # Optional: to focus on realistic range
plt.show()

# The plot is ledt-skewed thus most listsings are expensive

```

```python
sns.boxplot(x='room_type', y='price', data=df)
plt.title('Price Variation by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price')
plt.ylim(0, 500)
plt.show()

# the medians will indicate if there are overpriced listings according to the room type
```

```python
sns.scatterplot(x='number_of_reviews', y='price', data=df)
plt.title('Price vs Number of Reviews')
plt.xlabel('Number of Reviews')
plt.ylabel('Price')
plt.ylim(0, 500)
plt.show()

#Cheaper listings get more reviews, hence most preferrable
```

```python
sns.scatterplot(x='longitude', y='latitude', hue='room_type', data=df)
plt.title('Geographic Distribution of Listings by Room Type')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(loc='upper right')
plt.show()

# this shows most preffered location, which will help in setting up property ie. strategizing
```

```python
numeric_cols = df.select_dtypes(include=['float64', 'int64'])
corr = numeric_cols.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# the plot shows correlation of the features
```
