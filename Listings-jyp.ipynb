```python
# There are a few categories in the neighbourhhod column and room_type

from sklearn.preprocessing import LabelEncoder
enc1= LabelEncoder()
df['Neighbourhood_group_enc']= enc1.fit_transform(df['neighbourhood_group'])
df.drop(columns=['neighbourhood_group'],inplace= True)
df['room_type.enc']= enc1.fit_transform(df['room_type'])
df.drop(columns=['room_type'],inplace=True)
df.head()
```

```python
# For the Neighbourhood column and license, there are too many categories, of which i will use frequency encoder to encode it
freq= df['neighbourhood'].value_counts()
freq2=df['license'].value_counts()
df['neighbourhood_freq']= df['neighbourhood'].map(freq)
df['licence_freq']= df['license'].map(freq2)
df.drop(columns=[ 'neighbourhood','license','last_review'],inplace=True)
df.head()
```

# MACHINE LEARNING

    My goal is to predict house prices,henceforth i shall test using regression models 
    ~PROCEDURE:
    * Fixing the X and y variables
    * Split the data
    * Fitting standard scaler
    * Fitting the model
    * Hyperparameter tuning
    * Evaluating the model

```python
X= df.drop('price',axis=1)
y= df['price']

from sklearn.model_selection import train_test_split


X_train,X_test,y_train,y_test= train_test_split(X,y,random_state=3)

print("X_train shape: {}".format(X_train.shape))
print("y_train shape: {}".format(y_train.shape))

from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()

X_train_scaler=scaler.fit_transform(X_train)
X_test_scaler=scaler.fit(X_test)

from sklearn.linear_model import LinearRegression
lrg= LinearRegression(fit_intercept=True,n_jobs=-1)
'''Fitting the model'''

model= lrg.fit(X_train,y_train)
model
```

```python
# Evaluation

from sklearn.metrics import mean_squared_error,mean_absolute_error, r2_score

y_pred= model.predict(X_test)

print("\n--- Regression Model Evaluation ---")
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"MSE: {mse:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"MAE: {mae:.4f}")
print(f"R-squared: {r2:.4f}")
```
